````python
# Obtener bandwidth adecuado con KernelDensity()
def bandwidth_KernelD(datos: "recibe un array"):
    """Esta función regresa el bandwidth estimado con las librerías KernelDensity y GridSearchCV."""
    x = np.array(datos) # Se convierten a un arreglo los datos confirmados
    x = x.flatten() # Cambiamos el array a dimensión 1 x n
    grid = GridSearchCV(KernelDensity(), {"bandwidth": np.linspace(0.1, datos.max(), 200)}, cv = 20) # validación
    grid.fit(x[:, None])
    return grid.best_params_
x1 = np.array(data_confirmados).flatten()
bandwith_ker1 = bandwidth_KernelD(x1)
bandwith_ker1
````

````python
# Obtener la bandwidth con la función en lugar de un array con KDEMultivariate
def bandwidth_KDEM(datos: "debe ser un array"):
    """Esta función regresa el bandwidth estimado con la librería de KDEMultivariate."""
    kde_multivariate = KDEMultivariate(data = datos, var_type = "c")
    return kde_multivariate.bw
bandwidth_opt1 = bandwidth_KDEM(x1)
bandwidth_opt1
````

Se corre el siguiente código con ambos bandwidths para ver cual es el mejor. En este caso, el que se ve menos sobreajustado es el de la librería `GridSearchCV`. El número de bins se establece como el máximo de los datos sobre el bandwidth elegido debido a que la librería de `KernelDensity` calcula de esa manera el número de clases del histograma para poder construir la función de densidad. Así, el número de barras del histograma siempre concordará visualmente con la función de densidad creada.

````python
# Se grafica el histograma de los datos junto con la función de densidad obtenida
def hist_vs_sklearn(bandwidth: "bandwith obtenido con KDEMUltivariate", datos: "debe ser un array", 
                   titulo: "título de la gráfica generada", xlabel: "nombre del eje x"):
    """Esta función regresa una gráfica del histograma de los datos usando el KDEMultivariate junto con una función de densidad
    generada."""
    kde_sklearn = KernelDensity(bandwidth = bandwidth, kernel = "gaussian").fit(datos[:, np.newaxis])
    f_skle = lambda datos: np.exp(kde_sklearn.score_samples(datos[:, None])) # Función de densidad g(x)
    x_grid = np.arange(datos.min(), datos.max())
    plt.figure(figsize = [15, 7])
    plt.plot(x_grid, f_skle(x_grid))
    plt.hist(datos, bins = int(datos.max() / bandwidth), density = True, color = "g")
    plt.title(titulo, fontsize = 20)
    plt.xlabel(xlabel, fontsize = 12)
    plt.ylabel("Frecuencia relativa", fontsize = 12)
    plt.show()
    return f_skle, x_grid
hist1 = hist_vs_sklearn(bandwith_ker1["bandwidth"], x1, "Histograma de casos confirmados de Covid-19 en México", 
                        "Confirmados diarios")
hist1
````

A continuación se utiliza la librería de `Fitter` para poder obtener las distribuciones de `SciPy Stats` que mejor se ajustan a los datos. Decidimos correrla con todas las distribuciones posibles para no pasar por alto ninguna. El parámetro `timeout` de la función `Fitter.fitter` indica cuánto tiempo debe pasar para que la función omita esa distribución y se pase a la siguiente. Decidimos ponerlo de 10 segundos en lugar del valor default de 30 segundos para que se tarde menos en correr. Las advertencias indican cuales distribuciones fueron saltadas. El parámetro de `bins` se corrió con el número de bins que muestró el algoritmo anteior para que ambos métodos de modelado se basen en el mismo número de clases y la comparación sea más justa y menos sesgada. Así se hizo con el resto de variables.

````python
# Encontrar si es que hay, las primeras 3 distribuciones que mejor se ajusten a los datos
def fitter(data: "debe ser una array", bins: "número de clases"):
    """Esta función utiliza la librería Fitter para obtener las distribuciones que mejor se ajustan a los datos"""
    fitter1 = Fitter(data, bins = bins, timeout = 10, density = True)
    fitter1.fit()
    plt.figure(figsize = [15, 7])
    return fitter1
fitter1 = fitter(x1, 16)
fitter1.summary()
````

````python
# Gráficas cuantil cuantil
def graficas_qq(best: "las distribuciones que mejor se ajustaron", data: "debe ser un array", 
                fit: "lo que retorna la función de fitter"):
    """Esta función regresa las gráficas cuartil cuartil de las distribuciones dadas y los datos reales."""
    fig, axs = plt.subplots(1, 3, figsize = [18, 6])
    for i in range(len(best)):
        params = fit.fitted_param[best[i]]
        sm.qqplot(data, dist = getattr(st, best[i])(*params), line = "q", ax = axs[i])
        axs[i].set_title(f"Gráfica Q-Q de la distribución {best[i]}", fontsize = 15)
best1 = ["dweibull", "gennorm", "dgamma"]
graficas_qq(best1, x1, fitter1)
````

La siguiente función genera una variable aletoria mediante el método de aceptación y rechazo. El ciclo `while` que se encuentra dentro se agregó con el objetivo de que el número de valores de la variable aleatoria sea exactamente igual que el número de valores de los datos reales. Esto se hizo con el objetivo de optimizar y mejorar la simulación.

````python
def Acep_rechazo(R2: "Variables distruidas U ~ U(0, 1)", R1: "Variables distribuidas como g(x)",
                 f: "Función objetivo a generar", t: "Función que mayor a f", data: "debe ser un array"):
    """Esta función retorna la variable aleatoria generada con el método de aceptación y rechazo."""
    f_x = f(R1)
    t_x = t(R1)
    condition = R2 * t_x <= f_x
    cont = 0
    x = []
    i = 0
    while cont < len(data):
        if condition[i]:
            plt.plot(R1[i], R2[i] * t_x[i], "om")
            x.append(R1[i])
            cont += 1
        else:
            plt.plot(R1[i], R2[i] * t_x[i], "oc") 
        i += 1
    plt.show()
    return x
````

Después, ajustamos el método de aceptación y rechazo con la función de densidad obtenida anteriormente con la librería de `GridSearchCV`. Notamos que el método de `optimize.fmin` necesita un parámetro que le indique más o menos por dónde se encuentra el máximo en la distribución, especialmente si esta tiene más de un máximo local. Si no especificamos bien este parámetro, la función encontrará el máximo en el primer pico y no en el segundo. Por eso la función que construimos recibe un parámetro objetivo.

Adicionalmente, utilizamos un método de reducción de varianza con números complementarios. Sin embargo, este tiene una debilidad: debido a que la función anterior la definimos de tal manera que se acepten X número de puntos, entonces el número de iteraciones (N) dependerá de este. En otras palabras, el contador `i` de la función anterior son las N veces que se simula. Pero no podemos saber cuantas serán hasta que corramos la siguiente función, y para correr ésta se necesita N. En conclusión, se entraría en una especie de ciclo. Por lo tanto, decidimos definir una N a ojo únicamente viendo la gráfica de la distribución y tanteando más o menos cuantos puntos se simularán en total. Quizás se aceptarán más o menos el 25%, por lo que N es más o menos el número de muestras en los datos multiplizado por 4.

Si definimos una N más grande de lo necesario (por ejemplo, 5,000), entonces el método de reducción de varianza no hará nada porque la primera mitad de R2 (2500 números aleatorios) es la única que se tomará para la simulación, y la otra mitad (los 2500 números complementarios), no se tomarán.

````python
np.random.seed(4567)
def grafica_acep_rechazo(x_grid: "x_grid", f_skle: "función de densidad generada", objetivo: "número cercano al máximo",
                        xlabel: "título del eje x", titulo: "título de la gráfica"):
    """Esta función regresa una gráfica de los puntos aceptados y rechazados."""
    plt.figure(figsize = [15, 7])
    plt.title(titulo, fontsize = 20)
    plt.ylabel("Probabilidad", fontsize = 12)
    plt.xlabel(xlabel, fontsize = 12)
    xa = x_grid
    f = f_skle
    a = f(optimize.fmin(lambda xa: -f(xa), objetivo, disp = False))
    t = lambda xa: a * np.ones([len(xa)])
    plt.plot(xa, f(xa), label = "$f(x)$")
    plt.plot(xa, t(xa), label = "$t(x) = a$")
    plt.legend(loc = "best");
    return f, t
graf1 = grafica_acep_rechazo(hist1[1] , hist1[0], 6000, "Casos confirmados", 
                             "Gráfica de aceptación y rechazo para $f(x)$ de casos confirmados")
graf1

# Validación del método con reducción de varianza por números complementarios
N = 1200 # número de puntos a simular, creemos que aceptará aproximadamente 1200
r2_i = np.random.rand(int(N / 2)) # como estoy tomando t(x) constante solo es necesario generar valores aleatorios U ~ (0, 1)
r2_c = 1 - r2_i
R2 = np.concatenate([r2_i, r2_c])
R1 = np.random.uniform(x1.min(), x1.max(), size = N)
V_A1 = Acep_rechazo(R2, R1, graf1[0] , graf1[1], x1)
````

A continuación se hace la prueba de chi cuadrada para la función de densidad arrojada por `GridSearchCV` y las 3 arrojadas por `Fitter` con el objetivo de ver cuáles se ajustan y cuáles no. Se escogió 0.05 como nivel de significancia (alpha $\alpha$).
- $H_0$ (hipótesis nula): La distribución se ajusta a los datos reales.
- $H_a$ (hipótesis alternativa): La distribución no se ajusta a los datos reales.

Para todas las variables aplicaremos las mismas hipótesis.

````python
# Prueba chi-cuadrada
def prueba_chi2(data: "debe ser un array", best: "lista de las distribuciones que mejor se ajustan", 
                va: "variable aleatoria generada en el método de aceptación y rechazo"):
    """Esta función retorna los resultados de la prueba de Chi cuadrada."""
    observada_kernel = np.histogram(data, bins = (1 + int(np.log2(len(data)))), density = True)[0] # regla de Sturges
    esperada_kernel =  np.histogram(va, bins = (1 + int(np.log2(len(va)))), density = True)[0]
    print(f" Resultado de Chi2 con KDEMultivariate: {st.chisquare(f_obs = observada_kernel, f_exp = esperada_kernel, axis = 0)}"
         )
    for dist in best:
        observada = np.histogram(data, bins = (1 + int(np.log2(len(data)))), density = True)[0]
        params = getattr(st, dist).fit(data)
        rvs = getattr(st, dist)(*params).rvs(len(data))
        esperada =  np.histogram(rvs, bins = (1 + int(np.log2(len(rvs)))), density = True)[0]
        Chi_est = st.chisquare(f_obs = observada, f_exp = esperada, axis = 0)
        print(f"\n Resultado de Chi2 para la distribución {dist}: {Chi_est}")
chi2_1 = prueba_chi2(x1, best1, V_A1)
chi2_1
````

Los resultados de la ```p-value``` indican que se acepta la hipótesis nula para la distribución construida por `GridSearchCV`, `st.dweibull` y `st.gennorm` (porque es mayor que alpha) y se rechaza para la distribución `dgamma` (porque es menor que alpha). Esto significa que un 95% de confianza, las primeras 3 distribuciones se ajustan a los datos, pero para la última no hay datos suficientes para poder concluir lo mismo.

````python
# Prueba de Kolmogorov
def prueba_kolmogorov(data: "debe ser un array", best: "lista de las distribuciones que mejor se ajustan",
                     va: "variable aleatoria generada en el método de aceptación y rechazo"):
    """Esta función retorna los resultados de la prueba de Kolmogorov-Smirnov."""
    print(f" Resultado de Kolmogorv-Smirnov con KDEMultivariate: {st.kstest(data, va)}")
    for dist in best:
        params = getattr(st, dist).fit(data)
        kolmogorov = st.kstest(data, dist, args = params)
        print(f"\n Resultado de Kolmogorov-Smirnov para la distribución {dist}: {kolmogorov}")
prueba_kolmogorov(x1, best1, V_A1)
````

Los resultados indican que se acepta la hipótesis nula para las distribuciones de `GridSearchCV` y de `st.gennorm`, pero se rechaza para la distribución `st.dweibull` y `dgamma`. Decidimos hacer la simulación con el modelo de `GridSearchCV` debido a que este es el que arroja un ```p-value``` más alto para ambas pruebas. A continuación se muestra una comparación de histogramas entre los datos reales y los ajustados con esta distribución. A la función se le pueden meter tanto los datos de una distribución de `Scipy Stats` como los de la variable aleatoria generada con `GridSearchCV`. Dependiendo de cuál se haya escogido, la función ejecutará un condicional.

````python
# Comparación de histogramas
def histreal_vs_histva(data: "debe ser un array", titulo: "título de la gráfica", xlabel: "nombre del eje x",
                       bins: "número de clases", va: "variable aleatoria generada en el método de aceptación y rechazo" = None, 
                       dist: "distribución elegida" = None, fit: "lo que retorna la función fitter" = None):
    """Esta función retorna en una sóla gráfica el histograma con la distribución ajustada y con los datos reales con propósitos
    de comparación. Hay que tener en cuenta que si se le pasa una 'va' a la función, esto quiere decir que se elegió la función
    de distribución de KDEMultivariate, entonces no se le pasa un 'dist'. Si se le pasa un 'dist', quiere decir que se escogió
    una distribución convencional de la paquetería de SciPy Stats. Igualmente, si esta función recibe un 'dist', también un 'fit'."""
    if dist == None:
        plt.figure(figsize = [15, 7])
        plt.title(titulo, fontsize = 20)
        plt.hist(va, bins = bins, density = True, 
                 label = "Distribución ajustada", color = "black");
        plt.hist(data, bins = bins, density = True, label = "Distribución datos reales", color = "r");
        plt.xlabel(xlabel)
        plt.ylabel("Frecuencia relativa")
        plt.legend(loc = "best")
    if dist != None:
        params = fit.fitted_param[dist]
        rvs = getattr(st, dist)(*params).rvs(len(data))
        plt.figure(figsize = [15, 7])
        plt.title(titulo, fontsize = 20)
        plt.hist(rvs, bins = bins, density = True, 
                 label = "Distribución ajustada", color = "black");
        plt.hist(data, bins = bins, density = True, label = "Distribución datos reales", color = "r");
        plt.xlabel(xlabel)
        plt.ylabel("Frecuencia relativa")
        plt.legend(loc = "best")
        return rvs
histreal_vs_histva(x1, "Histogramas de distribución ajustada y datos reales de los casos confirmados", "Confirmados diarios", 
                   16, va = V_A1)
````

A continuación se construye una función que define la diferencia máxima que hay en los datos, con el objetivo de realizar una simulación montecarlo posteriormente y usar este resultado como parámetro que define el cambio de los datos.

````python
# Se establece un limite para tomar los datos consecutivos de la simulación
def lim_KDE(data: "debe ser un array"):
    c = data
    c = pd.DataFrame(c)
    dif = abs((c.shift() - c).dropna())
    limite = dif.max()[0].astype("int")
    return limite
limite1 = lim_KDE(x1)
limite1
````

Creamos una función que realiza la simulación de Montecarlo como se revisó en clase: iterativamente, el algoritmo escoge aleatoriamente un valor de las variables aleatorias generadas anteriormente, pero con una restricción: el límite definido acota el rango de valores que puede tomar la simulación. Esto con el objetivo de que no haya cambios tan bruscos, pues debido a que la variable misma se generó de manera uniformemente aleatoria, hay cambios muy bruscos entre los datos sin esta restricción (por ejemplo, la simulación podría pasar de tener 4000 casos confirmados a 27000 de un día para otro, lo cual no es realista).

Además, la función recibe el parámetro `r`, el cual es la tasa de movilidad. Decidimos usar este parámetro multiplicando los valores escogidos aleatoriamente de la variable aleatoria por `r`. De esta manera, un supuesto de la simulación es que la tasa de movilidad es directamente proporcional a las ocurrencias de los datos. Por ejemplo, si en lo que llevamos de pandemia hubiera habido una tasa de movilidad del 50%, los casos que hubiéramos visto hubieran sido la mitad de lo que realmente han sido.

Es importante recalcar que el primer valor de la simulación es el último de la serie real. Así que la simulación inicia en el día en el último día de la serie.

````python
# Se hace una función que simula 30 días tomando valores de la distribución dentro de los limites establecidos
np.random.seed(4567)
V_A1 = np.array(V_A1)
choice = np.random.choice
def mes(dias: "semanas del forecast", r: "tasa de movilidad", limite: "límite estimado", data: "debe ser un array",
        va: "variable aleatoria genera en el método de aceptación y rechazo"):
    l = []
    k = data[-1]
    l.append(k)
    for i in range(dias):
        z = k + limite
        y = k - limite
        k = choice(va[(y < va) & (va < z)]) * r
        l.append(k)
    return np.array(l)
````

Después, corremos la simulación con una fecha futura de 30 días, en 10,000 escenarios, con una tasa de movilidad del 100%.

````python
# Se hace una función que simula 30 días tomando valores de la distribución dentro de los limites establecidos
np.random.seed(4567)
escenarios, dias, r1 = 10000, 30, 1
confirmados_esperados1 = np.array(list(map(lambda x: mes(dias, r1, limite1, x1, np.array(V_A1)), range(escenarios)))).astype("int")
````

Después realizamos una función para graficar la series real junto con una sóla simulación (para que la gráfica sea agradable visualmente y no saturarla de escenarios), y con la media de todos ellos. También mostramos un `DataFrame` con los datos arrojados por la simulación. Se tuvo que convertir el índice anterior en formato de `Fecha`, a uno tipo `string` debido a que sólo de esa manera podíamos graficar todos estos objetos en una sóla figura. Además, esta función tiene la opción de meterle como parámetro el escenario que se quiere graficar (de los 10,000 que hay).

````python
# Graficar el forecast
def grafica_forecast(forecast: "un escenario", data: "debe ser DataFrame", titulo: "título de la gráfica", 
                     ylabel: "nombre del eje y", no_escenario: "número de escenario del total generado" = 0):
    simulacion = pd.DataFrame(forecast[no_escenario])
    simulacion.index = pd.to_datetime(simulacion.index, unit = "d", origin = pd.Timestamp(pd.to_datetime(data.index[-1])))
    simulacion.index = simulacion.index.astype("str")
    simulacion_media = pd.DataFrame(forecast.mean(axis = 0))
    simulacion_media.index = pd.to_datetime(simulacion_media.index, unit = "d", origin = pd.Timestamp(pd.to_datetime(data.index[-1])))
    simulacion_media.index = simulacion_media.index.astype("str")
    data.index = data.index.astype("str")
    plt.figure(figsize = [15, 7])
    plt.plot(data, color = "orange", label = "Datos reales")
    plt.plot(simulacion, color = "orange", label = "Ejemplo de un escenario de simulación", alpha = 0.5)
    plt.plot(simulacion_media, color = "pink", label = "Media de los datos simulados en los 10,000 escenarios")
    plt.xlabel("Días", fontsize = 12)
    plt.ylabel(ylabel, fontsize = 12)
    plt.title(titulo, fontsize = 20)
    plt.xticks(rotation = 75, ticks = np.arange(0, len(data) + len(simulacion), 15))
    plt.legend()
    plt.show()
    return simulacion_media.head(10)
sim1 = grafica_forecast(confirmados_esperados1, data_confirmados, 
                        "Datos reales contra datos simulados de los casos diarios de Covid-19 en México (cuarentena nula)",
                        "Confirmados diarios", 3333)
sim1
````

Finalmente, calculamos la cardinalidad de la variable en este nodo como una forma de medir el riesgo. Como se definió en el marco teórico de este proyecto, establecimos la cardinalidad comparando el último día de la simulación (de ahora en un mes) en base a los cuartiles de los datos.

````python
# Calcular cardinalidad de esta variable en este escenario
def classification(sim, hist):
    if sim < hist.quantile(0.25):
        cardinalidad = 1
    elif hist.quantile(0.25) <= sim < hist.median():
        cardinalidad = 2
    elif hist.median() <= sim < hist.quantile(0.75):
        cardinalidad = 3
    elif hist.quantile(0.75) <= sim:
        cardinalidad = 4
    return cardinalidad
classification(sim1.iloc[-1][0], data_confirmados["SMA7 Confirmados"])
````

En este caso se espera un repunte de los casos esperados, pasando de 3,800 a 5,400 aproximadamente, sin embargo, con una tendencia constante en este nivel de 5,500 casos. La clasificación que se le asigna a este comportamiento es de los casos confirmados esperados es de 2.

En la tabla se resultados puede notar que las 3 distribuciones que mejor se ajustan a los datos son: foldcauchy, genlogistic y cauchy ya que tienen el error cuadratico más bajo. Así mismo en el histograma se puede notar el ajuste de dichas funciones a los bisn correspondientes. 

Para poder escojer el modelo que mejor se ajusta a los datos es  necesario hacer las pruebas que se definieron en la pimera variable: Cuantil - Cuantil, Chi cuadrada y Kolmogorov - Smirnov.

Como se puede obsrvar la función que tiene un mayor ajuste relativamente hablando es la genlogistic pues la parte media de la distribución y los ultimos cuantiles tienen un ajuste considerablemente adecuado, sin embargo, los primeros cuantiles muestran una diferencia importante respecto a los datos originales. Graficamente se podrían descartar desde este punto las funciones foldcauchy y cauchy. Aún con lo anterior es importante realizar las pruebas de chi cuadrada y kolmogorov - smirnov.

Para continuar con dichas pruebas se realiza el algoritmo de aceptación y rechazo y así generar una variable aleatoria que distribuya como la función KDE que obtuvimos y que tenga la misma longitud que nuestros datos originales.
